```python
X = df[audio_features]
```

```python
X = X.dropna()
```

```python
# Scaling 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
# PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_scaled)
```

```python
# ==========================================
# PHASE 3: ELBOW METHOD
# ==========================================

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
inertia_values = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(list(k_range), inertia_values, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```

```python
# ==========================================
# PHASE 4: SILHOUETTE SCORE
# ==========================================

from sklearn.metrics import silhouette_score
```

```python
silhouette_scores = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    silhouette_scores.append(score)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(list(k_range), silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different k')
plt.grid(True)
plt.show()
```

```python
# Show scores clearly
k_score_df = pd.DataFrame({
    'k': list(k_range),
    'inertia': inertia_values,
    'silhouette_score': silhouette_scores
})

k_score_df
```

```python
df_model = df.copy()

df_model = df_model.dropna(subset=audio_features)

X = df_model[audio_features]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
# ==========================================
# PHASE 5: FINAL KMEANS MODEL
# ==========================================

best_k = 3   # after checking elbow/silhouette

final_kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
df_model['cluster'] = final_kmeans.fit_predict(X_scaled)

df_model[['cluster']].head()
```

```python
# ==========================================
# PHASE 6: PCA VISUALIZATION
# ==========================================

from sklearn.decomposition import PCA
import seaborn as sns
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

pca_df = pd.DataFrame(X_pca, columns=['PCA1', 'PCA2'], index=df_model.index)
pca_df['cluster'] = df_model['cluster'].values

print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Total explained variance:", pca.explained_variance_ratio_.sum())
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=pca_df,
    x='PCA1',
    y='PCA2',
    hue='cluster',
    palette='tab10',
    s=60
)
plt.title('Song Clusters Visualized with PCA')
plt.legend(title='Cluster')
plt.show()
```

```python
# ==========================================
# PHASE 7: CLUSTER PROFILING
# ==========================================

cluster_profile = df_model.groupby('cluster')[audio_features].mean().round(2)
cluster_profile
```

```python
# Size of each cluster
cluster_sizes = df_model['cluster'].value_counts().sort_index()
cluster_sizes
```

```python
plt.figure(figsize=(8, 5))
cluster_sizes.plot(kind='bar')
plt.xlabel('Cluster')
plt.ylabel('Number of Songs')
plt.title('Cluster Size Distribution')
plt.grid(axis='y')
plt.show()
```

```python
# ==========================================
# PHASE 8: HEATMAP OF CLUSTER MEANS
# ==========================================

plt.figure(figsize=(12, 6))
sns.heatmap(cluster_profile, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Average Audio Features per Cluster')
plt.show()
```

```python
# ==========================================
# PHASE 9: CLUSTER INTERPRETATION HELPERS
# ==========================================

def interpret_cluster(row):
    if row['energy'] >= 0.6 and row['danceability'] >= 0.6:
        return "High-energy, danceable tracks"
    elif row['acousticness'] >= 0.6 and row['energy'] <= 0.4:
        return "Chill / acoustic tracks"
    elif row['instrumentalness'] >= 0.5:
        return "Instrumental-focused tracks"
    elif row['speechiness'] >= 0.4:
        return "Speech-heavy tracks"
    else:
        return "Mixed/general tracks"

cluster_profile['cluster_description'] = cluster_profile.apply(interpret_cluster, axis=1)
cluster_profile
```

```python
# ==========================================
# PHASE 10: FINAL OUTPUT DATASET
# ==========================================

# Merge cluster labels back to the original filtered dataframe
final_df = df_model.copy()

final_df.head()
```

```python
# ==========================================
# PHASE 11: SAMPLE SONGS PER CLUSTER
# ==========================================

sample_columns = ['name_song', 'name_artists', 'cluster'] if 'name_song' in final_df.columns else ['cluster']
final_df[sample_columns].sort_values('cluster').head(20)
```

```python
for c in sorted(final_df['cluster'].unique()):
    print(f"\nCluster {c} sample songs:")
    display(final_df[final_df['cluster'] == c][sample_columns].head(5))
```

```python
# ==========================================
# PHASE 12: FINAL METRICS
# ==========================================

from sklearn.metrics import davies_bouldin_score
```

```python
sil_score = silhouette_score(X_scaled, final_df['cluster'])
db_index = davies_bouldin_score(X_scaled, final_df['cluster'])
inertia = final_kmeans.inertia_

print(f"Silhouette Score: {sil_score:.4f}")
print(f"Davies-Bouldin Index: {db_index:.4f}")
print(f"Inertia: {inertia:.4f}")
```
